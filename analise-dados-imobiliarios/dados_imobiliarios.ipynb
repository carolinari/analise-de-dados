# 1. Importe as bibliotecas.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('houses_to_rent_v3.txt')
df.head()
```

# 2. Crie uma visualização que mostre a distribuição dos valores de aluguel, 'rent amount (R$)', dos imóveis. Pode utilizar o gráfico que você julgar mais adequado. Qual faixa de valores de aluguel é mais comum?

```python
df['rent amount (R$)'].describe()
```

```python
# Histograma

plt.figure(figsize=(10, 5))
plt.title('Distribuição de valores de aluguel dos imóveis')
plt.hist(df['rent amount (R$)'],bins=60)
plt.show()
```

```python
# Gráfico de densidade

plt.figure(figsize=(10, 5))
plt.title('Distribuição de valores de aluguel dos imóveis')
sns.kdeplot(df['rent amount (R$)'], shade=True)
plt.show()
```

```python
# Gráfico de densidade com histograma

plt.figure(figsize=(10, 5))
plt.title('Distribuição de valores de aluguel dos imóveis')
sns.distplot(df['rent amount (R$)'])
plt.show()
```

A faixa de valores de aluguel mais comum está na área de 0 e 5000 reais. A média de valores de aluguel é aproximadamente 2741 reais.

# 3. Crie uma visualização que mostre a distribuição dos valores de seguro contra incêndio, 'fire insurance (R$)', dos imóveis. Existem valores incomuns?

```python
df['fire insurance (R$)'].describe()
```

```python
# Histograma

plt.figure(figsize=(10, 5))
plt.title('Distribuição de valores de seguro contra incêndio dos imóveis')
plt.hist(df['fire insurance (R$)'],bins=45)
plt.show()
```

```python
# Gráfico de densidade

plt.figure(figsize=(10, 5))
plt.title('Distribuição de valores de seguro contra incêndio dos imóveis')
sns.kdeplot(df['fire insurance (R$)'], shade=True)
plt.show()
```

```python
# Gráfico de densidade com histograma

plt.figure(figsize=(10, 5))
plt.title('Distribuição de valores de seguro contra incêndio dos imóveis')
sns.distplot(df['fire insurance (R$)'])
plt.show()
```

Sim, existem valores incomuns. O maior valor registrado dos valores de seguro contra incêndio, 'fire insurance (R$)', dos imóveis é de 305 reais, enquanto a faixa de valores mais comum está na área de 0 a 50 reais. A média é de aproximadamente 36 reais.
