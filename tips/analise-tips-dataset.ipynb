# Análise exploratória de dados | Tips Dataset Analysis

O conjunto de dados é composto por 244 linhas e 7 variáveis que representam alguns dados de gorjetas registrados.
* Total da conta - bill in dollars,
* Gorjeta - tip in dollars,
* Sexo - sex of the bill payer,
* Fumantes e não fumantes - whether there were smokers in the party,
* Dia - day of the week,
* Horário - time of day,
* Tamanho da festa - size of the party.

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

pairplot irá traçar distribuições entre pares em todo o DataFrame (para as colunas numéricas) e suporta um argumento de matiz de cor (para colunas categóricas)

```python
sns.pairplot(tips)
```

```python
sns.heatmap(tips.corr(),annot=True)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

- Apresentam maior correlaçao total_bill e tip.

# Valor total da conta (total_bill) ao longo dos dias (day)

```python
sns.countplot(x='day',data=tips, palette="Set1")
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day', palette="Set1")
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, palette="Set1")
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, palette="Set1")
```

Os dias registrados no conjunto de dados são: Quinta-feira, sexta-feira, sábado e domingo (Thur, Fri, Sat e Sun).

* As sextas-feiras são os dias com menos clientes.
* Há mais clientes no fim de semana, os sábados são os dias mais movimentados, seguidos pelos domingos.

* A média do valor total da conta é mais baixo na sexta-feira.
* A média do valor total da conta aumeta no final de semana.

Nas sextas-feiras há menos clientes e, no geral, a média dos valores totais das contas é menor que os demais.

# Valor total da conta (total_bill) e Gênero (sex)

```python
# contando o número de ocorrências

sns.countplot(x='sex',data=tips, palette="Set1")
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', palette="Set1")
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex', palette="Set1")
```

```python
sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips, palette="Set1", split=True)
```

```python
sns.boxplot(x="day", y="total_bill", hue="sex",data=tips, palette="Set1")
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',split=True,palette='Set1')
```

```python
sns.countplot(x='day', hue='sex',data=tips, palette="Set1")
```

* O conjunto de dados apresenta mais registros de clientes homens.
* A média do valor total das contas dos homens e das mulheres aos sábado são semelhantes.
* Nas quintas-feiras a quantidade de mulheres que pagam o total da conta é maior que a quantidade de homens.
* A quantidade de mulheres e homens que pagam a conta às quintas e sextas-feiras é quase igual. 
* Na sexta-feira e no final de semana, a quantidade de homens que pagam as contas é superior a quantidade de mulheres.

# Valor total da conta (total_bill) e Fumantes e não fumantes (smoker)

```python
sns.countplot(x='time', hue='smoker',data=tips, palette="Set1")
```

```python
sns.countplot(x='day', hue='smoker',data=tips, palette="Set1")
```

* Na maioria dos dias a quantidade de não fumantes é superior a quantidade de fumantes.
* Apenas às sextas-feiras a quantidade de clientes fumantes é superior a quantidade de clientes não fumantes. 
* A quantidade de não fumantes é maior que o dobro da quantidade de fumantes às quintas-feiras e aos domingos.
